# 2. Outliers detection on weather_data & filtered_weather_data

```python
def detect_global_outliers(df, column, method="iqr", threshold=1.5, z_threshold=3):
    """
    Detect outliers considering cities as one.
    
    Parameters:
        method: "iqr" (interquartile interval) or "zscore" (std).
        threshold: IQR (default: 1.5).
        z_threshold: Z-Score (default: 3).
    """
    data = df[column].copy()
    
    if method == "iqr":
        Q1 = data.quantile(0.25)
        Q3 = data.quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - threshold * IQR
        upper = Q3 + threshold * IQR
        outliers = df[(data < lower) | (data > upper)]
        
    elif method == "zscore":
        z_scores = np.abs(stats.zscore(data.dropna()))
        outliers = df[z_scores > z_threshold]
    
    return outliers
```

```python
# List of variables to analyze
print("Outliers for weather_data")
variables = ['T2M', 'RH2M', 'ALLSKY_SFC_SW_DWN']

# Dictionary to store results
outliers_global = {}

print("For IQR threshold = 1.5")
for var in variables:
    outliers = detect_global_outliers(filtered_weather_data, var, method="iqr")
    outliers_global[var] = outliers
    print(f"Outliers in {var}: {len(outliers)} ({(len(outliers)/len(filtered_weather_data))*100:.2f}%)")
    
print()

print("For IQR threshold = 2")
for var in variables:
    outliers = detect_global_outliers(filtered_weather_data, var, threshold=2.0,  method="iqr")
    outliers_global[var] = outliers
    print(f"Outliers in {var}: {len(outliers)} ({(len(outliers)/len(filtered_weather_data))*100:.2f}%)")

print()

print("For IQR threshold = 3")
for var in variables:
    outliers = detect_global_outliers(filtered_weather_data, var, threshold=3.0, method="iqr")
    outliers_global[var] = outliers
    print(f"Outliers in {var}: {len(outliers)} ({(len(outliers)/len(filtered_weather_data))*100:.2f}%)")
```

# 4. Feature Generation for Vegetative & Reproductive Stages

# Temperature (T2M) Features

- **Days with T2M in the ideal range during vegetative stage**  
  `[T2M_days_ideal_veg]`

- **Days with T2M in the ideal range during reproductive stage**  
  `[T2M_days_ideal_rep]`

- **Days with T2M below the lower threshold during vegetative stage**  
  `[T2M_days_low_veg]`

- **Days with T2M below the lower threshold during reproductive stage**  
  `[T2M_days_low_rep]`

- **Days with extremely low T2M during vegetative stage**  
  `[T2M_days_extremely_low_veg]`

- **Days with extremely low T2M during reproductive stage**  
  `[T2M_days_extremely_low_rep]`

> High temperatures have very low frequency in both stages, so they are unlikely to contribute meaningful patterns for modeling.  
> However, **extremely low temperatures** can significantly impact growth and should be considered as important stress indicators.

```python
# --- 1) Parameters and mappings ---
# Months for each growth stage
vegetative_months = [5, 6]   # May, June
reproductive_months = [7, 8] # July, August

# Temperature thresholds (in °C)
# Vegetative stage
veg_low_thresh = 18
veg_extremely_low_thresh = 14
veg_ideal_min = 18
veg_ideal_max = 30

# Reproductive stage
rep_low_thresh = 22
rep_extremely_low_thresh = 18
rep_ideal_min = 22
rep_ideal_max = 30


# --- 2) Add boolean indicator columns ---
df = filtered_weather_data.copy()  # Your DataFrame must contain 'month', 'T2M', '市町村名', and 'year'

# Vegetative stage
df['is_ideal_vegetative'] = (
    df['month'].isin(vegetative_months) &
    df['T2M'].between(veg_ideal_min, veg_ideal_max)
)
df['is_low_vegetative'] = (
    df['month'].isin(vegetative_months) &
    (df['T2M'] < veg_low_thresh) & 
    (df['T2M'] >= veg_extremely_low_thresh)
)
df['is_extremely_low_vegetative'] = (
    df['month'].isin(vegetative_months) &
    (df['T2M'] < veg_extremely_low_thresh)
)

# Reproductive stage
df['is_ideal_reproductive'] = (
    df['month'].isin(reproductive_months) &
    df['T2M'].between(rep_ideal_min, rep_ideal_max)
)
df['is_low_reproductive'] = (
    df['month'].isin(reproductive_months) &
    (df['T2M'] < rep_low_thresh) & 
    (df['T2M'] >= rep_extremely_low_thresh)
)
df['is_extremely_low_reproductive'] = (
    df['month'].isin(reproductive_months) &
    (df['T2M'] < rep_extremely_low_thresh)
)


# --- 3) Aggregate counts by city and year ---
t2m_features = (
    df
    .groupby(['市町村名', 'year'])
    .agg({
        'is_ideal_vegetative'          : 'sum',
        'is_low_vegetative'            : 'sum',
        'is_extremely_low_vegetative'  : 'sum',
        'is_ideal_reproductive'        : 'sum',
        'is_low_reproductive'          : 'sum',
        'is_extremely_low_reproductive': 'sum',
    })
    .rename(columns={
        'is_ideal_vegetative'          : 'T2M_days_ideal_vegetative',
        'is_low_vegetative'            : 'T2M_days_low_vegetative',
        'is_extremely_low_vegetative'  : 'T2M_days_extremely_low_vegetative',
        'is_ideal_reproductive'        : 'T2M_days_ideal_reproductive',
        'is_low_reproductive'          : 'T2M_days_low_reproductive',
        'is_extremely_low_reproductive': 'T2M_days_extremely_low_reproductive',
    })
    .reset_index()
)

# --- 4) Monthly mean temperature from May (5) to August (8) ---
monthly_means = (
    df[df['month'].between(5, 8)]
    .groupby(['市町村名', 'year', 'month'])['T2M']
    .mean()
    .unstack(fill_value=np.nan)  # columns become months
    .rename(columns={
        5: 'T2M_5',
        6: 'T2M_6',
        7: 'T2M_7',
        8: 'T2M_8'
    })
    .reset_index()
)

# --- 5) Merge everything ---
t2m_features = t2m_features.merge(monthly_means, on=['市町村名', 'year'], how='left')

t2m_features.columns
```

# Humidity (RH2M) Features

- **Days with RH2M in the ideal range during vegetative stage**  
  `[RH2M_days_ideal_veg]`

- **Days with RH2M in the ideal range during reproductive stage**  
  `[RH2M_days_ideal_rep]`

- **Days with RH2M between 80% and 90% during vegetative stage (high humidity)**  
  `[RH2M_days_high_veg]`

- **Days with RH2M between 80% and 90% during reproductive stage (high humidity)**  
  `[RH2M_days_high_rep]`

- **Days with RH2M above 90% during vegetative stage (very high humidity)**  
  `[RH2M_days_very_high_veg]`

- **Days with RH2M above 90% during reproductive stage (very high humidity)**  
  `[RH2M_days_very_high_rep]`

```python
# --- 1) Parameters ---
# Growth stage months
vegetative_months = [5, 6]
reproductive_months = [7, 8]

# RH2M thresholds (%)
rh2m_low_thresh = 60
rh2m_high_thresh = 80
rh2m_very_high_thresh = 90  # Extreme humidity

# --- 2) Add boolean indicator columns ---
df = filtered_weather_data.copy()

# Vegetative stage
df['is_rh2m_ideal_veg'] = (
    df['month'].isin(vegetative_months) &
    df['RH2M'].between(rh2m_low_thresh, rh2m_high_thresh)
)
df['is_rh2m_high_veg'] = (
    df['month'].isin(vegetative_months) &
    df['RH2M'].gt(rh2m_high_thresh) & df['RH2M'].le(rh2m_very_high_thresh)
)
df['is_rh2m_very_high_veg'] = (
    df['month'].isin(vegetative_months) &
    df['RH2M'].gt(rh2m_very_high_thresh)
)

# Reproductive stage
df['is_rh2m_ideal_rep'] = (
    df['month'].isin(reproductive_months) &
    df['RH2M'].between(rh2m_low_thresh, rh2m_high_thresh)
)
df['is_rh2m_high_rep'] = (
    df['month'].isin(reproductive_months) &
    df['RH2M'].gt(rh2m_high_thresh) & df['RH2M'].le(rh2m_very_high_thresh)
)
df['is_rh2m_very_high_rep'] = (
    df['month'].isin(reproductive_months) &
    df['RH2M'].gt(rh2m_very_high_thresh)
)

# --- 3) Aggregate counts ---
rh2m_features = (
    df
    .groupby(['市町村名', 'year'])
    .agg({
        'is_rh2m_ideal_veg'     : 'sum',
        'is_rh2m_high_veg'      : 'sum',
        'is_rh2m_very_high_veg' : 'sum',
        'is_rh2m_ideal_rep'     : 'sum',
        'is_rh2m_high_rep'      : 'sum',
        'is_rh2m_very_high_rep' : 'sum',
    })
    .rename(columns={
        'is_rh2m_ideal_veg'     : 'RH2M_days_ideal_veg',
        'is_rh2m_high_veg'      : 'RH2M_days_high_veg',
        'is_rh2m_very_high_veg' : 'RH2M_days_very_high_veg',
        'is_rh2m_ideal_rep'     : 'RH2M_days_ideal_rep',
        'is_rh2m_high_rep'      : 'RH2M_days_high_rep',
        'is_rh2m_very_high_rep' : 'RH2M_days_very_high_rep',
    })
    .reset_index()
)

# Final result
display(rh2m_features)
```

# **Solar Radiation (ALLSKY_SFC_SW_DWN) features**  
  - Days with ALLSKY in the ideal range during vegetative stage  
    `[ALLSKY_days_ideal_veg]`  
  - Days with ALLSKY in the ideal range during reproductive stage  
    `[ALLSKY_days_ideal_rep]`  
  - Days with ALLSKY below the lower threshold during vegetative stage  
    `[ALLSKY_days_low_veg]`  
  - Days with ALLSKY below the lower threshold during reproductive stage  
    `[ALLSKY_days_low_rep]`
  - ALLSKY mean per month  
_______________________________

```python
# --- 1) Parameters ---
# Months for each growth stage
vegetative_months = [5, 6]
reproductive_months = [7, 8]

# Solar radiation thresholds (example values in MJ/m²/day)
allsky_extremely_low_thresh = 10
allsky_low_thresh = 15
allsky_high_thresh = 25

# --- 2) Add boolean indicator columns ---
df = filtered_weather_data.copy()

# Vegetative stage
df['is_allsky_ideal_veg'] = (
    df['month'].isin(vegetative_months) &
    df['ALLSKY_SFC_SW_DWN'].between(allsky_low_thresh, allsky_high_thresh)
)
df['is_allsky_low_veg'] = (
    df['month'].isin(vegetative_months) &
    (df['ALLSKY_SFC_SW_DWN'] < allsky_low_thresh) &
    (df['ALLSKY_SFC_SW_DWN'] >= allsky_extremely_low_thresh)
)
df['is_allsky_extremely_low_veg'] = (
    df['month'].isin(vegetative_months) &
    (df['ALLSKY_SFC_SW_DWN'] < allsky_extremely_low_thresh)
)
df['is_allsky_high_veg'] = (
    df['month'].isin(vegetative_months) &
    (df['ALLSKY_SFC_SW_DWN'] > allsky_high_thresh)
)

# Reproductive stage
df['is_allsky_ideal_rep'] = (
    df['month'].isin(reproductive_months) &
    df['ALLSKY_SFC_SW_DWN'].between(allsky_low_thresh, allsky_high_thresh)
)
df['is_allsky_low_rep'] = (
    df['month'].isin(reproductive_months) &
    (df['ALLSKY_SFC_SW_DWN'] < allsky_low_thresh) &
    (df['ALLSKY_SFC_SW_DWN'] >= allsky_extremely_low_thresh)
)
df['is_allsky_extremely_low_rep'] = (
    df['month'].isin(reproductive_months) &
    (df['ALLSKY_SFC_SW_DWN'] < allsky_extremely_low_thresh)
)
df['is_allsky_high_rep'] = (
    df['month'].isin(reproductive_months) &
    (df['ALLSKY_SFC_SW_DWN'] > allsky_high_thresh)
)

# --- 2.5) Add mean ALLSKY per month (5, 6, 7, 8) ---
monthly_allsky = (
    df[df['month'].isin([5, 6, 7, 8])]
    .groupby(['市町村名', 'year', 'month'])['ALLSKY_SFC_SW_DWN']
    .mean()
    .unstack(fill_value=0)
    .rename(columns={5: 'ALLSKY_5', 6: 'ALLSKY_6', 7: 'ALLSKY_7', 8: 'ALLSKY_8'})
    .reset_index()
)

# --- 3) Aggregate counts by city and year ---
allsky_features = (
    df
    .groupby(['市町村名', 'year'])
    .agg({
        'is_allsky_ideal_veg': 'sum',
        'is_allsky_low_veg': 'sum',
        'is_allsky_extremely_low_veg': 'sum',
        'is_allsky_high_veg': 'sum',
        'is_allsky_ideal_rep': 'sum',
        'is_allsky_low_rep': 'sum',
        'is_allsky_extremely_low_rep': 'sum',
        'is_allsky_high_rep': 'sum',
    })
    .rename(columns={
        'is_allsky_ideal_veg': 'ALLSKY_days_ideal_veg',
        'is_allsky_low_veg': 'ALLSKY_days_low_veg',
        'is_allsky_extremely_low_veg': 'ALLSKY_days_extremely_low_veg',
        'is_allsky_high_veg': 'ALLSKY_days_high_veg',
        'is_allsky_ideal_rep': 'ALLSKY_days_ideal_rep',
        'is_allsky_low_rep': 'ALLSKY_days_low_rep',
        'is_allsky_extremely_low_rep': 'ALLSKY_days_extremely_low_rep',
        'is_allsky_high_rep': 'ALLSKY_days_high_rep',
    })
    .reset_index()
)

# --- 4) Merge monthly means into final dataset ---
allsky_features = allsky_features.merge(monthly_allsky, on=['市町村名', 'year'], how='left')

# Show the resulting DataFrame
display(allsky_features)
```

# **Combined Ideal Conditions**  
  - Days when all three variables are simultaneously ideal during vegetative stage  
    `[ideal_combo_days_veg]`  
  - Days when all three variables are simultaneously ideal during reproductive stage  
    `[ideal_combo_days_rep]`

```python
# --- 1) Define stage months ---
vegetative_months = [5, 6]
reproductive_months = [7, 8]

# --- 2) Define ideal thresholds ---
# Temperature
t2m_ideal_veg_min = 18
t2m_ideal_veg_max = 30
t2m_ideal_rep_min = 22
t2m_ideal_rep_max = 30

# Humidity
rh2m_ideal_min = 60
rh2m_ideal_max = 80

# Solar Radiation (MJ/m²/day)
allsky_ideal_min = 15
allsky_ideal_max = 25

# --- 3) Define boolean columns for ideal combinations ---

# Vegetative stage
df['is_ideal_combo_veg'] = (
    df['month'].isin(vegetative_months) &
    df['T2M'].between(t2m_ideal_veg_min, t2m_ideal_veg_max) &
    df['RH2M'].between(rh2m_ideal_min, rh2m_ideal_max) &
    df['ALLSKY_SFC_SW_DWN'].between(allsky_ideal_min, allsky_ideal_max)
)

# Reproductive stage
df['is_ideal_combo_rep'] = (
    df['month'].isin(reproductive_months) &
    df['T2M'].between(t2m_ideal_rep_min, t2m_ideal_rep_max) &
    df['RH2M'].between(rh2m_ideal_min, rh2m_ideal_max) &
    df['ALLSKY_SFC_SW_DWN'].between(allsky_ideal_min, allsky_ideal_max)
)

# --- 4) Aggregate by city and year ---
ideal_combo_features = (
    df
    .groupby(['市町村名', 'year'])
    .agg({
        'is_ideal_combo_veg': 'sum',
        'is_ideal_combo_rep': 'sum',
    })
    .rename(columns={
        'is_ideal_combo_veg': 'ideal_combo_days_veg',
        'is_ideal_combo_rep': 'ideal_combo_days_rep',
    })
    .reset_index()
)

# Display result
display(ideal_combo_features)
```

# **Combined Stress Conditions**
    - Days when any major stress combination occurs (high T2M + low RH2M + low ALLSKY) during vegetative stage  
    `[stress_combo_days_veg]`  
    - Days when any major stress combination occurs during reproductive stage  
    `[stress_combo_days_rep]` 

```python
# --- 1) Parameters ---
vegetative_months = [5, 6]
reproductive_months = [7, 8]

extreme_low_t2m_thresh_veg = 14
extreme_low_t2m_thresh_rep = 18
extreme_high_rh2m_thresh = 90
extreme_low_allsky_thresh = 10

# --- 2) Add stress condition columns ---
df = filtered_weather_data.copy()

# Vegetative stage triple stress
df['is_stress_combo_veg'] = (
    df['month'].isin(vegetative_months) &
    (df['T2M'] < extreme_low_t2m_thresh_veg) &
    (df['RH2M'] > extreme_high_rh2m_thresh) &
    (df['ALLSKY_SFC_SW_DWN'] < extreme_low_allsky_thresh)
)

# Reproductive stage triple stress
df['is_stress_combo_rep'] = (
    df['month'].isin(reproductive_months) &
    (df['T2M'] < extreme_low_t2m_thresh_rep) &
    (df['RH2M'] > extreme_high_rh2m_thresh) &
    (df['ALLSKY_SFC_SW_DWN'] < extreme_low_allsky_thresh)
)

# --- 3) Aggregate counts ---
stress_combo_features = (
    df
    .groupby(['市町村名', 'year'])
    .agg({
        'is_stress_combo_veg': 'sum',
        'is_stress_combo_rep': 'sum'
    })
    .rename(columns={
        'is_stress_combo_veg': 'stress_combo_days_veg',
        'is_stress_combo_rep': 'stress_combo_days_rep'
    })
    .reset_index()
)

display(stress_combo_features)
```

# **Growing Degree Days (GDD)**
GDD (Growing Degree Days) is a weather-based indicator used to estimate how much heat is available for plants or insects to grow on a given day.

Simple Explanation: <br>
Every plant or insect species needs a certain amount of warmth to develop.<br>
GDD measures the heat units they receive daily.<br>

It’s calculated using the daily average temperature, compared to a base temperature (usually the minimum needed for growth, like 10°C for many crops).
  - Cumulative GDD over the vegetative stage (base 10°C for rice)  
    `[GDD_sum_veg]`  
  - Cumulative GDD over the reproductive stage (base 10°C for rice)  
    `[GDD_sum_rep]` 
________________________

```python
# --- 1) Parameters ---
vegetative_months = [5, 6]   # May and June
reproductive_months = [7, 8] # July and August
gdd_base_temp = 10           # Base temperature for rice (°C)

# --- 2) Copy original DataFrame ---
df = filtered_weather_data.copy()

# --- 3) Calculate daily GDD ---
df['GDD'] = (df['T2M'] - gdd_base_temp).clip(lower=0)

# --- 4) Stage-specific GDD allocation ---
df['GDD_veg'] = df.apply(lambda row: row['GDD'] if row['month'] in vegetative_months else 0, axis=1)
df['GDD_rep'] = df.apply(lambda row: row['GDD'] if row['month'] in reproductive_months else 0, axis=1)

# --- 5) Aggregate GDD by municipality and year ---
gdd_features = (
    df
    .groupby(['市町村名', 'year'])
    .agg({
        'GDD_veg': 'sum',
        'GDD_rep': 'sum'
    })
    .rename(columns={
        'GDD_veg': 'GDD_sum_veg',
        'GDD_rep': 'GDD_sum_rep'
    })
    .reset_index()
)

display(gdd_features)
```

# VPD (Vapor Pressure Deficit)
VPD (Vapor Pressure Deficit) is a weather indicator that tells us how "thirsty" the air is — or in other words, how much moisture the air wants to pull from the rice plants through evaporation and transpiration.

🌾 Simple Explanation (for rice farming):<br>
Plants release water into the air through tiny pores (stomata) — this process is called transpiration.<br>
VPD measures the drying power of the air.<br>
Low VPD = humid air, less water loss<br>
High VPD = dry air, more water pulled from the plant (can cause stress)
_______________________

```python
# --- 1) Parameters ---
vegetative_months = [5, 6]   # May and June
reproductive_months = [7, 8] # July and August

# --- 2) Copy original DataFrame ---
df = filtered_weather_data.copy()

# --- 3) Calculate daily VPD ---
# Saturation vapor pressure (es) using Tetens formula
df['es'] = 0.6108 * np.exp((17.27 * df['T2M']) / (df['T2M'] + 237.3))

# Actual vapor pressure (ea)
df['ea'] = df['es'] * (df['RH2M'] / 100)

# VPD in kPa
df['VPD'] = df['es'] - df['ea']

# --- 4) Stage-specific VPD allocation ---
df['VPD_veg'] = df.apply(lambda row: row['VPD'] if row['month'] in vegetative_months else 0, axis=1)
df['VPD_rep'] = df.apply(lambda row: row['VPD'] if row['month'] in reproductive_months else 0, axis=1)

# --- 5) Aggregate VPD by municipality and year ---
vpd_features = (
    df
    .groupby(['市町村名', 'year'])
    .agg({
        'VPD_veg': 'sum',
        'VPD_rep': 'sum'
    })
    .rename(columns={
        'VPD_veg': 'VPD_sum_veg',
        'VPD_rep': 'VPD_sum_rep'
    })
    .reset_index()
)

display(vpd_features)
```

# ET₀ (Reference Evapotranspiration)
ET₀ (Reference Evapotranspiration) is a weather-based indicator that estimates how much water a reference crop (like well-watered grass) would lose to the atmosphere through evaporation and transpiration under current weather conditions.

🌾 Simple Explanation (for rice farming):<br>
ET₀ tells us how much water the weather "wants" to pull from the field — it combines temperature, humidity, radiation, and sometimes wind.<br>
It’s like a weather-driven water demand meter.

High ET₀ = hot, sunny, dry weather → more water needed by rice plants

Low ET₀ = cool, cloudy, humid weather → less water needed<br>
We can use accumulated ET₀ during the growing season to estimate water demand or drought pressure.
________________________

```python
# --- 1) Parameters ---
vegetative_months = [5, 6]
reproductive_months = [7, 8]
k = 0.0023
T_range = 10  # Assumed average temperature range (Tmax - Tmin)

# --- 2) Copy original DataFrame ---
df = filtered_weather_data.copy()

# --- 3) Estimate ET₀ using simplified Hargreaves-Samani ---
df['ET0'] = k * (df['T2M'] + 17.8) * (T_range ** 0.5) * df['ALLSKY_SFC_SW_DWN']

# --- 4) Stage-specific ET₀ allocation ---
df['ET0_veg'] = df.apply(lambda row: row['ET0'] if row['month'] in vegetative_months else 0, axis=1)
df['ET0_rep'] = df.apply(lambda row: row['ET0'] if row['month'] in reproductive_months else 0, axis=1)

# --- 5) Aggregate ET₀ by municipality and year ---
et0_features = (
    df
    .groupby(['市町村名', 'year'])
    .agg({
        'ET0_veg': 'sum',
        'ET0_rep': 'sum'
    })
    .rename(columns={
        'ET0_veg': 'ET0_sum_veg',
        'ET0_rep': 'ET0_sum_rep'
    })
    .reset_index()
)

display(et0_features)
```

# PTI (Photothermal Index)
PTI (Photothermal Index) combines temperature and sunlight to estimate how favorable the environment is for plant development — especially for flowering and grain filling stages.

🌾 Simple Explanation (for rice farming):<br>
PTI = Temperature × Solar Radiation<br>
Rice needs both warmth and sunlight to grow well.<br>
If it’s warm but cloudy, or sunny but cold, development slows.<br>
A high PTI means ideal conditions: warm and sunny → faster and healthier crop development.<br>

High PTI = great conditions for rice growth

Low PTI = limited growth due to either low sunlight or low temperature
____________________________

```python
# --- 1) Parameters ---
vegetative_months = [5, 6]   # May and June
reproductive_months = [7, 8] # July and August
gdd_base_temp = 10           # Base temperature for rice (°C)

# --- 2) Copy original DataFrame ---
df = filtered_weather_data.copy()

# --- 3) Calculate daily GDD ---
df['GDD'] = (df['T2M'] - gdd_base_temp).clip(lower=0)

# --- 4) Calculate daily PTI (Photothermal Index) ---
df['PTI'] = df['GDD'] / df['ALLSKY_SFC_SW_DWN']

# --- 5) Stage-specific PTI allocation ---
df['PTI_veg'] = df.apply(lambda row: row['PTI'] if row['month'] in vegetative_months else 0, axis=1)
df['PTI_rep'] = df.apply(lambda row: row['PTI'] if row['month'] in reproductive_months else 0, axis=1)

# --- 6) Aggregate PTI by municipality and year ---
pti_features = (
    df
    .groupby(['市町村名', 'year'])
    .agg({
        'PTI_veg': 'sum',
        'PTI_rep': 'sum'
    })
    .rename(columns={
        'PTI_veg': 'PTI_sum_veg',
        'PTI_rep': 'PTI_sum_rep'
    })
    .reset_index()
)

display(pti_features)
```

# PTU (Photothermal Unit)
PTU is similar to PTI, but it accumulates the effect over time — just like GDD. It’s the sum of daily PTI values across the season.

🌾 Simple Explanation (for rice farming):<br>
PTU adds up how much "growth energy" rice receives day by day from temperature and sunlight.<br>
It helps estimate how much progress the crop has made in its development.<br>

High PTU over the season = fast and consistent development

Low PTU = slower growth, possibly delayed harvest

___________________________

```python
# --- 1) Parameters ---
vegetative_months = [5, 6]   # May and June
reproductive_months = [7, 8] # July and August
gdd_base_temp = 10           # Base temperature for rice (°C)

# --- 2) Copy original DataFrame ---
df = filtered_weather_data.copy()

# --- 3) Calculate daily GDD ---
df['GDD'] = (df['T2M'] - gdd_base_temp).clip(lower=0)

# --- 4) Calculate daily PTU (Photo-Thermal Unit = GDD × Radiation) ---
df['PTU'] = df['GDD'] * df['ALLSKY_SFC_SW_DWN']

# --- 5) Stage-specific PTU allocation ---
df['PTU_veg'] = df.apply(lambda row: row['PTU'] if row['month'] in vegetative_months else 0, axis=1)
df['PTU_rep'] = df.apply(lambda row: row['PTU'] if row['month'] in reproductive_months else 0, axis=1)

# --- 6) Aggregate PTU by municipality and year ---
ptu_features = (
    df
    .groupby(['市町村名', 'year'])
    .agg({
        'PTU_veg': 'sum',
        'PTU_rep': 'sum'
    })
    .rename(columns={
        'PTU_veg': 'PTU_sum_veg',
        'PTU_rep': 'PTU_sum_rep'
    })
    .reset_index()
)

display(ptu_features)
```
